```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_ps.csv', sep=',')

melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)

melb_df['AgeBuilding'] = melb_df['Date'].dt.year - melb_df['YearBuilt']

melb_df = melb_df.drop('YearBuilt', axis=1)

melb_df['WeekdaySale'] = melb_df['Date'].dt.dayofweek

def get_weekend(weekday): 
    if weekday == 5 or weekday == 6:
        return 1
    else:
        return 0 
    
melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)

# display(melb_df[melb_df['Weekend'] == 1]['Price'].mean()  )

popular_agency = melb_df['SellerG'].value_counts().nlargest(49).index

# print(popular_agency)

melb_df['SellerG'] = melb_df['SellerG'].apply(lambda x: x if x in popular_agency else 'other')

# display(melb_df['SellerG'].value_counts())

Nelson_min = melb_df[ melb_df['SellerG'] == 'Nelson' ]['Price'].min()
other_min = melb_df[ melb_df['SellerG'] == 'other' ]['Price'].min()

print(Nelson_min)
print(other_min)

print(Nelson_min/other_min)
```

```python
import pandas as pd

sample = pd.Series(data = ['Опыт работы 8 лет 3 месяца', 'Опыт работы 1 год 2 месяца', 'Опыт работы 6 лет'])

def get_experience(arg):
   
    arg_list = arg.split(' ')
    
    total_month = 0
    
    if arg_list[-1].startswith('мес'):
        total_month += int(arg_list[-2])
    if arg_list[-1].startswith('лет') or arg_list[-1].startswith('го'):
        total_month += (int(arg_list[-2])*12)
    if arg_list[-3].startswith('лет') or arg_list[-3].startswith('го'):
        try:
            total_month += (int(arg_list[-4])*12)
        except:
            pass

    
    return total_month

 

sample.apply(get_experience)
```

```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_ps.csv', sep=',')

unique_list = []

for col in melb_df.columns:
    item = (col, melb_df[col].nunique(), melb_df[col].dtypes)
    unique_list.append(item)
    
unique_counts = pd.DataFrame(unique_list, columns=['Column_name', 'Num_unique', 'Type']).sort_values(by='Num_unique', ignore_index=True)

# display(unique_counts)


cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

max_unique_count = 150

for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')


# display(melb_df.info())

# print(melb_df['Regionname'].cat.categories)

melb_df['Type'] = melb_df['Type'].cat.rename_categories( {
    'u':'unit',
    'h':'house',
    't':'townhouse'
    } )

display(melb_df['Type'])
```
