```python
import pandas as pd 

melb_df_orig = pd.read_csv('data/melb_data_fe.csv', sep=',')

melb_df = melb_df_orig.copy()

melb_df['Date'] = pd.to_datetime(melb_df['Date'])

def quarter_ident(arg):
    if 1 <= arg < 4:
        return 1
    if 4 <= arg < 7:
        return 2
    if 7 <= arg < 10:
        return 3
    if 10 <= arg <= 12:
        return 4 
    
melb_df['Quarter'] = melb_df['Date'].dt.month.apply(quarter_ident)

cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']

for col in melb_df.columns:
    if melb_df[col].nunique() < 150 and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
        
melb_df.info()
    
```

```python
# TASK 6-3

import pandas as pd
import os


def concat_user_files(path):
    
    files_list = os.listdir(path)
    files_list.sort()
    
    df_files_users = pd.DataFrame()
    
    for csv_file in files_list:
        temp_path = path + '/' + csv_file
        temp_df = pd.read_csv(temp_path, sep=',')
        df_files_users = pd.concat(
        [df_files_users, temp_df],
        )
    df_files_users =df_files_users.drop_duplicates(ignore_index=True)
    return df_files_users
```
