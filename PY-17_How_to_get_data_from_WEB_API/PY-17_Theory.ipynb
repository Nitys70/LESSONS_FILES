**СБОР НЕСКОЛЬКИХ ЭЛЕМЕНТОВ: СОБИРАЕМ ВСЕ ССЫЛКИ НА СТРАНИЦЕ**

```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_programming_languages'

response = requests.get(url)

page = BeautifulSoup(response.text, 'html.parser')

print(page.find('a'))

links = page.find_all('a')

print(len(links))
print(len(set(links)))

dupl_list = []
trash_list = []

for link in links:
    if link in trash_list: 
        dupl_list.append(link)
    else:
        trash_list.append(link)


print([link.text for link in links[500:510]])

print(len(dupl_list))
```

**sleep**

```python
import requests
import time

token = 'a958c670a958c670a958c670b8aa400d4eaa958a958c670cf2d16c62fd96fd2afe01564'

url = 'https://api.vk.com/method/groups.getMembers'

count = 1000
offset = 0
user_ids = []

while offset < 5000:
    params = {'group_id': 'vk', 'v': 5.95, 'count': count, 'offset': offset, 'access_token': token}
    
    response = requests.get(url, params=params)
    data = response.json()
    
    user_ids += data['response']['items']
    
    offset += count
    
    print('Ожидаю 0.5 сек')
    time.sleep(0.5)
    
print('Цикл завершен, offset =', offset)
```
