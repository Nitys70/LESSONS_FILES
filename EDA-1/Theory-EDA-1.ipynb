```python
import pandas as pd
import matplotlib
import seaborn as sns
import numpy as np

data_reg = []

df = pd.read_csv('data/wine.csv')

for n in range(df.shape[0]):
    counter = 0
    
    if counter == 10:
        break 
    
    if (df.loc[n]['region_1'] != df.loc[n]['region_2']) and (df.loc[n]['region_2'] is not np.nan):
        data_reg.append([df.loc[n]['region_1'], df.loc[n]['region_2']])
        counter +=1


new_df = pd.DataFrame(data=data_reg, columns=['reg_1', 'reg_2'])

new_df[new_df['reg_1'] == 'Oregon']

new_df.to_csv('data/new_df.csv')

reg_dif_df = pd.read_csv('data/new_df.csv')
```
