```python
a = 'Опыт работы 10 месяцев Август 2018 — по насто...'


def exprns(elem):
    try:
        counter = 0
        elem = elem.split(' ')
        if ('го' or 'ле') in elem[3]:
            counter += int(elem[2])*12
            print('1 ', counter)
            print(elem[3])
        
        if 'мес' in elem[3]:
            counter += int(elem[2])
            print('2 ', counter)
        
        if 'мес' in elem[5]:
            counter += int(elem[4])
            print('3 ', counter)
        
        return counter
    except:
        return None

exprns(a)
```
