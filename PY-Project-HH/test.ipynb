```python
import pandas as pd

df = pd.read_csv('data/dst-3.0_16_1_hh_database.csv', sep=';')

# display(df.head())

def education_class(elem):
    elem=elem.lower()
    if 'неоконченное высшее' in elem:
        elem = 'неоконченное высшее'
    elif 'высшее' in elem:
        elem = 'высшее'
    if 'среднее специальное' in elem:
        elem = 'среднее специальное'
    elif 'среднее' in elem:
        elem = 'среднее'    
    return elem

df['Образование'] = df['Образование и ВУЗ'].apply(education_class)

# display(df[df['Образование'] == 'среднее'].head(5))
df['Образование']=df['Образование'].astype('category')
# df.info()
# df.info()
# df['Образование'].cat.categories

df['Пол'] = df['Пол, возраст'].apply(
    lambda x: 'М' if 'Мужчина' in x else 'Ж'
    )

df['Пол'].value_counts()

df['Возраст'] = df['Пол, возраст'].apply(lambda x: int(x.split(' ')[3]) )
df.head()

def exprns(elem):
    try:
        counter = 0
        elem = elem.split(' ')
        if 'го' or 'ле' in elem[3]:
            counter += int(elem[2])*12
        
        if 'мес' in elem[3]:
            counter += int(elem[2])
        
        if 'мес' in elem[5]:
            counter += int(elem[4])
        
        return counter
    
    except:
        return None
    
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(exprns)

df[df['Опыт работы (месяц)'] == 'NaN'].head()          



def city_filling(elem):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 
                  'Казань', 'Челябинск', 'Омск', 'Самара', 
                  'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 
                  'Воронеж', 'Волгоград' 
                  ]
    
    elem = elem.split(',')
    city = ''

    if 'Москва' in elem[0] or 'Санкт-Петер' in elem[0]:
        city = elem[0]
    elif elem[0] in million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'
        
    return city

df['Город'] = df['Город, переезд, командировки'].apply(city_filling)

df.head()


# Функция для вычленения элемента тематики перезда и его обработка

def relocate(elem):
    elem = elem.split(',')
    item_relocate = 'переезд'

    for item_relocate in elem:
        if item_relocate in elem:
            if 'не' in elem:
                return False
            else:
                return True
            

# Функция для вычленения элемента тематики командировки и его обработка,

def bis_trip(elem):
    elem = elem.split(',')
    item_bis_trip = 'коман'

    if item_bis_trip in elem[-1]:
        if 'не' in elem[-1]:
            return False
        else:
            return True 
    else:
        return False 


df['Готовность к переезду'] = df['Город, переезд, командировки'
                                 ].apply(relocate)
df['Готовность к командировкам'] = df['Город, переезд, командировки'
                                      ].apply(bis_trip)


df.head()

```
