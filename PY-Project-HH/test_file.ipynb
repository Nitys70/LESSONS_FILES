```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

hh_cv_orig = pd.read_csv('data/dst-3.0_16_1_hh_database.csv', sep=';')
hh_cv = hh_cv_orig.copy()



pd.set_option('display.max_rows', None)

# display(hh_cv[hh_cv['Опыт работы (месяц)'].isna()])

def exprns(elem):
    try:
     
        counter = 0
        elem = elem.split(' ')
            
        if 'го' in elem[3]:
            counter += int(elem[2])*12
            
        if 'лет' in elem[3]:
            counter += int(elem[2])*12
        
        if 'мес' in elem[3]:
            counter += int(elem[2])
        
        if 'меся' in elem[5]:
            counter += int(elem[4])
        return counter
    except: 
        return None

# Задание нового признака DF и удаление ненужного
   
hh_cv['Опыт работы (месяц)'] = hh_cv['Опыт работы'].apply(exprns)

# hh_cv = hh_cv.drop(['Опыт работы'], axis=1)

print('Медианное значение опыта: ', hh_cv['Опыт работы (месяц)'].median())
# print(hh_cv['Опыт работы (месяц)'].info())

test_text = hh_cv['Опыт работы'][1550]
test_text_list = test_text.split(' ')
print(test_text_list)
if 'ле' in test_text_list[3]:
    print('Функция работает')
number_mnth = int(test_text_list[2])*12
print('колво месяцев: ',number_mnth)

# null_cols = hh_cv.isnull().sum()

# nuul_2 = hh_cv[null_cols>0]

# nuul_2
display(hh_cv.iloc[1550:1551])

hh_cv.info()
```
