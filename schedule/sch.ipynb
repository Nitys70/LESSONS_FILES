```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

# Создание DF из исходного графика в формате эксель
scd_df = pd.read_excel('data/test_xl.xlsx')


date_keys = [(13, 'jan'), (15, 'feb'), (17, 'mar'), 
             (20, 'apr'), (22, 'may'), (24, 'jun'), 
             (28, 'july'), (30, 'aug'), (32, 'sep'), 
             (35, 'oct'), (37, 'nov'), (39, 'dec')
             ]

year_schdl_df = pd.DataFrame(columns = [list(range(1, 32))])

for elem in date_keys:
    mnth_arr_orig = (pd.DataFrame(scd_df.iloc[elem[0]]))[1:32].values
    mnth_arr = mnth_arr_orig.reshape((1, 31))
    mnth_arr_df = pd.DataFrame(
    data = mnth_arr, 
    columns = [list(range(1, 32))], 
    index = [elem[1]]  
    ) 
    year_schdl_df = pd.concat( [year_schdl_df, mnth_arr_df] )

def data_conv(elem):
    
    # print(elem.iloc[1])
    # print(elem)
    
    for n in range(0,12):
        print(elem.iloc[n])
        try:
            elem = int(elem.iloc[n])
            return elem
        except:
           
            if '/' in elem.iloc[n]:
                elem = elem.iloc[n].replace('/', ' ')
                elem = elem.split(' ')
                d = int(elem[0]) + int(elem[1])*0.4
                return d
            if elem.iloc[n] == 'В':
                return 0
            else:
                return elem*2


year_schdl_df[1] = year_schdl_df[1].apply(data_conv)
# year_schdl_df[1] = year_schdl_df[1].apply(lambda x: 'yr' if '/' in x else x)


display(year_schdl_df)
```
