### 4-4  Сколько уникальных тегов представлено в наборе данных?

```python
import re 

reg = re.compile('[^A-Z a-z 0-9]')

tag_set = set()

for index in range(df.shape[0]):
    tag_str = df.iloc[index]['tags']
    tag_list = list(tag_str.split(',')) 
    
    for elem in tag_list:
        tag_set.add( (reg.sub('',elem)).strip())

print(len(tag_set))    
```

### 4-5 Какой тег представлен в наибольшем числе отзывов?

```python
import re 

reg = re.compile('[^A-Z a-z 0-9]')

tag_dict = {}

for index in range(df.shape[0]):
    tag_str = df.iloc[index]['tags']
    tag_list = list(tag_str.split(',')) 
    
    for elem in tag_list:
        elem_cleared = (reg.sub('',elem)).strip()
        
        if elem_cleared in tag_dict.keys():
            tag_dict[f'{elem_cleared}'] += 1
        else:
            tag_dict[f'{elem_cleared}'] = 1
```

```python

tag_df = pd.DataFrame.from_dict(tag_dict, orient='index', columns=['qnt'])

# tag_df.head()

tag_dict

tag_df = tag_df.reset_index()
tag_df = tag_df.rename(columns={'index':'tag'})
tag_df = tag_df.sort_values(by='qnt', ascending=False)
tag_df
```
