```python
import pandas as pd

df = pd.read_excel('data/nys2025xslx.xlsx')
```

```python
month_list = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
df = df.rename(columns={'СОГЛАСОВАНО': 'month'})
```

```python
def month_convert(month):
    try:
        return month.strip()
    except:
        return month
df['month'] = df['month'].apply(month_convert)
```

```python
df_new = df[df.month.isin(month_list)]
```
